I'll be using Tensorflow as backend for keras in this notebook.

Both validation and test data consist of 8 images of horses and 8 images of humans each.

We'll be using ImageDataGenerator class from the keras preprocessing package to read and process images as this method helps load and process on images simultaneously. We don't have to store the images separately as such and then process on them.

```python
# For listing directory content, for paths etc.
import os

# For basic operations related to dataframe and matrices
import numpy as np
import pandas as pd

# For plotting operations
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread as mpimread
from matplotlib import rcParams
import seaborn as sns
from PIL import Image
%matplotlib inline

# For evaluation
from sklearn.metrics import confusion_matrix
```

```python
# For building CNN architecture

# For backend of keras 
import tensorflow as tf

# For building a feed-forward Neural Network
from keras.models import Sequential

# Import Conv2D layer - For building a convolutional layer (Since images are 2D hence 2D)
# Import MaxPooling2D layer - To perform subsampling using maxpooling for dim reduction
# Import Dense layer - For Fully Connected layers
# Import Dropout layer - For Regularization
# Import flatten layer - To flatten the input to fc layer after a convolutional layer
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten

# To use transfer learning from VGG 16
from keras.applications.vgg16 import VGG16

# In order to assign weights when a model is initialized
from keras import initializers

# In order to specify the algorithm for backpropogation
from keras.optimizers import SGD

# For Generation of images
from keras.preprocessing.image import ImageDataGenerator
```

```python
# To ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Define some constants
image_width = 224
image_height = 224
b_size = 4
num_epochs = 20
```

We will go chronologically in the following order:

- Image Preparation
- VGG 16 Model Tuning
- Model Fitting
- Model Evaluation 
- Test Prediction

# Image Preparation

```python
Train_path = os.getcwd() + '\\data\\Train'
Test_path = os.getcwd() + '\\data\\Test'
Val_path = os.getcwd() + '\\data\\Validation'
```

The `flow_from_directory` method of the `ImageDataGenerator` class can help us read images on the fly and feed them forward in the neural network.

Use this method to define train, test and validation sets.

```python
TrainSet = ImageDataGenerator().flow_from_directory(Train_path, target_size = (image_width, image_height), 
                                                    classes = ['Horses', 'Humans'], batch_size = b_size)

ValSet = ImageDataGenerator().flow_from_directory(Val_path, target_size = (image_width, image_height), 
                                                    classes = ['Horses', 'Humans'], batch_size = b_size)

TestSet = ImageDataGenerator().flow_from_directory(Test_path, target_size = (image_width, image_height), 
                                                    classes = ['Horses', 'Humans'], batch_size = b_size)
```

Just as a sanity check we can view a few images from our train, validation and test sets each.

```python
# The "next" method operates over an iterator to iterate
# Over a specific number of times (here batch_size) and returns the objects which the iterator
# which the iterator holds.

def plot_image_from_generator(gen):
    im, lb = next(gen)
    fig, axs = plt.subplots(1, gen.batch_size, figsize=(10,10))
#     rcParams['figure.figsize'] = (5,5)
    for i in range(gen.batch_size):
#         sub_plt = fig.add_subplot(1, gen.batch_size, i + 1)
        axs[i].axis('Off')
        axs[i].set_title(lb[i])
        axs[i].imshow(im[i].astype('uint8'))
```

```python
# From Training Data Generator
plot_image_from_generator(TrainSet)
```

```python
# From Validation Data Generator
plot_image_from_generator(ValSet)
```

```python
# From Test Data Generator
plot_image_from_generator(TestSet)
```

# Model Architecture

**VGG16 Architecture**:

In this architecture, we have a set of five almost identical sandwiched layers. Each sandwich is a series of 2 or 3 Convolutional 2D layers followed by a MaxPooling layer to subsample. There are five such sandwiches arranged together. As we go deeper, the number of kernels required to better classify goes on increasing. This is natural because the kernels deep down are learning from a combination of features from the previous layer. Since with each layer, number of features extracted increases, therefore the search space of number of features to combine and look into becomes exponentially high (in theory) but we limit it because practise, we don't observe those many combinations.

Then eventually, we have three fully connected layers and finally the output layer. Each fc layer has 4096 neurons. and eventually there's only two neurons to classify whether the image is that of a horse or a human.

The model has learned on a humongous motley of data in order to be able to classify an image belonging to one of a certain 1000 classes. [Refer the ImageNet Challenge](http://www.image-net.org/challenges/LSVRC/). That's why it's weights are in a very good local state and have figured out a lot of correlation between portions of images across many classes. We're going to use these weights for running our model and we won't alter them along the way (i.e. during training). This method of using architecture and parameters of some pretested/predefined model to match our needs is called **Transfer Learning**. 

```python
from keras.applications.vgg16 import VGG16

# Instantiate the VGG16 model
vgg_model = VGG16(weights = 'imagenet')

# Build a Sequential model
model = Sequential()

# Add all the layers of VGG16 model except the last layer to our model
# Our purpose is binary classification. VGG16 was trained to distinguish between 10 classes.
# Hence our output layer must be changed.

for layer in vgg_model.layers[:-1]:
    model.add(layer)

# Since these weights in the model are in quite a good state
# We don't want to train/update these in backpropagation. Hence,
# set the trainable parameter of these layers to be False.

for layer in model.layers:
    layer.trainable = False

# Add a dense layer at end for our binary classification 
model.add(Dense(2, activation = 'softmax', name = 'output'))

# have a look at the model architecture
print(model.summary())
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = SGD(lr = 0.05), metrics = ['accuracy'])
```

# Fitting the model

We have created the generator objects in order to read images on the fly and feed them forward in the neural network. Now, we can make use of those objects and fit our model using the fit_generator method in keras

**fit_generator():**

- generator - The generator object from which to sample the images
- steps_per_epoch - Total number of batches to extract from the given generator object per epoch.
eg. If batch_size = 4 and steps_per_epoch = 10, then the network will train on 4 * 10 = 40 images per epoch. i.e. Network will train on batch_size * steps_per_epoch number of samples in every epoch.
- epochs - Number of epochs for training the data
- validation_data - It can be a generator object or a tuple (x_validation, y_validation)
- validation_steps - If your validation_data is a generator, then how many validation_steps i.e. steps * batch_size number of samples to validate your model on.
- validation_freq - If you've provided the validation data, then this number specifies the number of epochs after which you've to validate i.e. if validation_freq = 5, then after every 5 epochs, validation will happen.
- shuffle - Whether to shuffle the batches at the beginning of each epoch.

```python
model.fit_generator(generator = TrainSet, steps_per_epoch = 8,
                   epochs = num_epochs, verbose = 1, 
                   validation_data = ValSet, validation_steps= 2,
                   validation_freq = 1, shuffle = True)
```

# Model Evaluation

**evaluate_generator**:

Evaluates the model on a data generator.

- generator: A generator object
- steps: Number of steps per epoch. If your batch size is n, then the evaluation will occur on steps * n samples.
- verbose: 0,1,2 for nothing, detailed and short descriptions respectively

Let us evaluate the model on validation dataset

```python
loss, accuracy = model.evaluate_generator(generator = ValSet, steps = 6, verbose = 0)
print(f"The validation loss is {np.round(loss,2)} and the accuracy on validation set is {np.round(accuracy, 3)}")
```

```python
predictions = model.predict_generator(generator = ValSet)
```

# Image Augmentation

We saw no substantial improvement in the accuracy even after we used VGG16 over our previous simple CNN model. Let us try to augment the images and use a bigger batch_size to enhance our training even further. Let's compare the performance after that.

The `ImageDataGenerator` class provided in the preprocessing module of keras helps us to create images by using a lot of operators such as `rotation_range`, `horizontal_flip` etc. For a complete list of operations, check out the [documentation page here](https://keras.io/preprocessing/image/#imagedatagenerator-class)

```python
b_size = 8

AugDataGen = ImageDataGenerator(rotation_range = 10,
                            width_shift_range = 0.1,
                            brightness_range = (0.5,1),
                            shear_range = 0.1,
                            zoom_range = (0.5,2),
                            horizontal_flip = True,
                            vertical_flip= True)

TrainAugData = AugDataGen.flow_from_directory(Train_path, target_size = (image_width, image_height), 
                                            classes = ['Horses', 'Humans'], batch_size = b_size)

ValAugData = AugDataGen.flow_from_directory(Val_path, target_size = (image_width, image_height), 
                                            classes = ['Horses', 'Humans'], batch_size = b_size)
```

After augmentation, we can see there's a stretch/rotation/flip, brightness has been changed, image is rotated and zoomed out etc. even when images are sampled from the same directory. The process of augmentation, essentially creates new data based on the given data so that the neural network can learn better.  

```python
plot_image_from_generator(TrainAugData)
```

```python
plot_image_from_generator(TrainSet)
```

 **Word of Caution:** Natural images can be augmented and our network will benefit from learning more such images but in applications which are sensitive to orientation or color or brightness etc. augmentation will not be helpful. For instance, if we were to do an alphabet recognition and if a train image is a b, then vertical flip will make it a p, horizontal flip will make it a d and both together will make it a q. It's therefore essential to check if augmentation doesn't hamper with the task at hand i.e. classification.
 
 Let's use this new augmented data and train the model with this data

```python
num_epochs = 10
model.fit_generator(generator = TrainAugData, steps_per_epoch = 8,
                   epochs = num_epochs, verbose = 1, 
                   validation_data = ValAugData, validation_steps= 2,
                   validation_freq = 1, shuffle = True)
```

```python
test_predictions = model.predict_generator(TestSet)
y_test_pred = np.argmax(test_predictions, axis = 1)
```

```python
print(y_test_pred)
print(TestSet.labels)
```

```python
cm = confusion_matrix(y_test, y_test_pred)
data = pd.DataFrame(cm, index = labels, columns = labels)
sns.heatmap(data, annot = True, cmap = 'YlGnBu')
plt.xticks(rotation = 90); plt.yticks(rotation = 0)
plt.show()
```

```python
plot_image_from_generator(TestSet)
```
