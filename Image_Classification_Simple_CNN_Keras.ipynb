Once, we know this, then if we collect more data, that will help our model learn better anyways. Eventually, I'll also demonstrate how we can make use of preprocessing library of keras to generate new images from a given set of images.

```python
# For listing directory content, for paths etc.
import os

# For basic operations related to dataframe and matrices
import numpy as np
import pandas as pd

# For plotting operations
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread as mpimread
from matplotlib import rcParams
import seaborn as sns
from PIL import Image
%matplotlib inline

# For evaluation
from sklearn.metrics import confusion_matrix
```

```python
# For building CNN architecture

# For backend of keras 
import tensorflow as tf

# For building a feed-forward Neural Network
from keras.models import Sequential

# Import Conv2D layer - For building a convolutional layer (Since images are 2D hence 2D)
# Import MaxPooling2D layer - To perform subsampling using maxpooling for dim reduction
# Import Dense layer - For Fully Connected layers
# Import Dropout layer - For Regularization
# Import flatten layer - To flatten the input to fc layer after a convolutional layer
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten

# In order to assign weights when a model is initialized
from keras import initializers

# In order to specify the algorithm for backpropogation
from keras.optimizers import SGD

# To one hot encode the classes
from keras.utils import to_categorical

# Generation of images
from keras.preprocessing.image import ImageDataGenerator
```

```python
# To ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Define some constants
image_width = 224
image_height = 224
b_size = 4
num_epochs = 10
```

We will go chronologically in the following order:

- Image Preparation
- Model Definition
- Model Fitting
- Model Evaluation 
- Test Prediction

# Image Preparation

```python
Train_path = os.getcwd() + '\\data\\Train'
Test_path = os.getcwd() + '\\data\\Test'
Val_path = os.getcwd() + '\\data\\Validation'
```

The `flow_from_directory` method of the `ImageDataGenerator` class can help us read images on the fly and feed them forward in the neural network.

Use this method to define train, test and validation sets.

```python
TrainSet = ImageDataGenerator().flow_from_directory(Train_path, target_size = (image_width, image_height), 
                                                    classes = ['Horses', 'Humans'], batch_size = b_size)

ValSet = ImageDataGenerator().flow_from_directory(Val_path, target_size = (image_width, image_height), 
                                                    classes = ['Horses', 'Humans'], batch_size = b_size)

TestSet = ImageDataGenerator().flow_from_directory(Test_path, target_size = (image_width, image_height), 
                                                    classes = ['Horses', 'Humans'], batch_size = b_size)
```

Just as a sanity check we can view a few images from our train, validation and test sets each.

```python
# The "next" method operates over an iterator to iterate
# Over a specific number of times (here batch_size) and returns the objects which the iterator
# which the iterator holds.

def plot_image_from_generator(gen):
    im, lb = next(gen)
    fig = plt.figure()
    for i in range(b_size):
        sub_plt = fig.add_subplot(1, b_size, i + 1)
        sub_plt.axis('Off')
        sub_plt.set_title(lb[i])
        plt.imshow(im[i].astype('uint8'))
```

```python
# From Training Data Generator
plot_image_from_generator(TrainSet)
```

```python
# From Validation Data Generator
plot_image_from_generator(ValSet)
```

```python
# From Test Data Generator
plot_image_from_generator(TestSet)
```

# Model Architecture

**Convolutional Layer Parameters:**
- filters - Number of kernels you wish to have in a layer.
- kernel size - width x height of the kernel
- strides - horizontal sweep x vertical sweep of the kernel across the given image
- padding - valid means edges won't be considered and same means edges will be considered by padding 0s along the sides/corners.

*The parameters below also apply to a dense layer*
- activation - Activation function
- input_shape - If first layer is convolutional, only then this comes into picture. The size of the input image is to be mentioned here.
- use_bias - Whether to use bias or not; Default is True.
- kernel_initializer - What weights to assign to the kernel when the model is initialized
- bias_initializer - What values to assign to the biases on initialization

**MaxPooling Layer Parameters:**
- pool_size - An integer or a tuple of integers specifying the pooling window size.
- strides - If None (Not specified), it will assume the pool_size and continue to subsample without overlap. Else, one can manually specify.

```python
model = Sequential([
    Conv2D(filters = 32, kernel_size = (3, 3), strides = (1, 1), padding = 'valid', 
          activation = 'relu', input_shape = (image_width, image_height, 3), 
          kernel_initializer = initializers.RandomNormal(0, 0.5),
          bias_initializer = initializers.Zeros(), use_bias = True),
    
    MaxPooling2D(pool_size = (2,2), strides = 2),
    
    Conv2D(32, (3,3), strides = (1,1), activation = 'relu'),
    
    MaxPooling2D(pool_size = (2,2), strides = 2),
    
    Flatten(),
    
    Dense(32, activation = 'relu'),
    
    Dropout(0.2),
    
    Dense(2, activation = 'softmax')
])
```

```python
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = SGD(lr = 0.05), metrics = ['accuracy'])
```

# Fitting the model

We have created the generator objects in order to read images on the fly and feed them forward in the neural network. Now, we can make use of those objects and fit our model using the fit_generator method in keras

**fit_generator():**

- generator - The generator object from which to sample the images
- steps_per_epoch - Total number of batches to extract from the given generator object per epoch.
eg. If batch_size = 4 and steps_per_epoch = 10, then the network will train on 4 * 10 = 40 images per epoch. i.e. Network will train on batch_size * steps_per_epoch number of samples in every epoch.
- epochs - Number of epochs for training the data
- validation_data - It can be a generator object or a tuple (x_validation, y_validation)
- validation_steps - If your validation_data is a generator, then how many validation_steps i.e. steps * batch_size number of samples to validate your model on.
- validation_freq - If you've provided the validation data, then this number specifies the number of epochs after which you've to validate i.e. if validation_freq = 5, then after every 5 epochs, validation will happen.
- shuffle - Whether to shuffle the batches at the beginning of each epoch.

```python
model.fit_generator(generator = TrainSet, steps_per_epoch = 20,
                   epochs = num_epochs, verbose = 1, 
                   validation_data = ValSet, validation_steps= 2,
                   validation_freq = 1, shuffle = True)
```

# Model Evaluation

**evaluate_generator**:

Evaluates the model on a data generator.

- generator: A generator object
- steps: Number of steps per epoch. If your batch size is n, then the evaluation will occur on steps * n samples.
- verbose: 0,1,2 for nothing, detailed and short descriptions respectively

```python
loss, accuracy = model.evaluate_generator(generator = ValSet, steps = 6, verbose = 0)
print(f"The validation loss is {np.round(loss,2)} and the accuracy on validation set is {np.round(accuracy, 3)}")
```

```python
predictions = model.predict_generator(generator = ValSet)
```

# Sidenote: Image Augmentation

We have discussed above how a simple CNN Network can be built using keras. In order to data augmentation, we can specify multiple parameters inside the ImageDataGenerator class which can achieve the desired effect. Let me illustrate this as a sidenote below using a separate ImageDataGenerator class. 

I will only be using a limited number of techniques to augment images. For a comprehensive list of all possible operations, please refer [this link.](https://keras.io/preprocessing/image/#imagedatagenerator-class)

```python
AugDataGen = ImageDataGenerator(rotation_range = 10,
                            width_shift_range = 0.1,
                            brightness_range = (0.5,1),
                            zoom_range = (0.5,2),
                            horizontal_flip = True,
                            vertical_flip= True)

AugData = AugDataGen.flow_from_directory(Train_path, target_size = (image_width, image_height), 
                                                    classes = ['Horses', 'Humans'], batch_size = b_size)
```

After augmentation, we can see there's a stretch/rotation/flip, brightness has been changed, image is rotated and zoomed out etc. even when images are sampled from the same directory. The process of augmentation, essentially creates new data based on the given data so that the neural network can learn better.  

```python
plot_image_from_generator(AugData)
```

```python
plot_image_from_generator(TrainSet)
```

 **Word of Caution:** Natural images can be augmented and our network will benefit from learning more such images but in applications which are sensitive to orientation or color or brightness etc. augmentation will not be helpful. For instance, if we were to do an alphabet recognition and if a train image is a b, then vertical flip will make it a p, horizontal flip will make it a d and both together will make it a q. It's therefore essential to check if augmentation doesn't hamper with the task at hand i.e. classification.
